##Завдання 1
Попросити користувача ввести рядок (* реалізувати введення та обробку та виведення будь якої кількості рядків +1 бал)
До рядків є наступні вимоги:
*  кожен рядок має складатися з 3 частин, розділених комами
*  перша частина має містити число з плаваючою "крапкою"
*  друга та третя частини можуть містити лише літери та пробіли

Якщо рядок не задовільняє вимогам, його не потрібно виводити.

Необхідно вивести отримані дані в наступному вигляді:
*  Стовпчик 1 = число має бути відформатоване до 2-х знаків після коми та розміщене в першому стовпчику з вирівнюванням по правому краю (ширина стовпчика - 7)
*  рядки тексту мають бути очищені від зайвих пробілів на початку та в кінці
*  Стовпчик 2 = перший з рядків має бути відформатований як речення та розміщений по центру (ширина стовпчика - 20)
*  Стовпчик 3 = другий з рядків має бути в верхньому регістрі та вирівняний за лівим краєм (ширина стовпчика - 20)

Між стовпчиками та по краям мають бути розділювачі на ваш вибір.

```python
#Місце для коду Завдання 1
# Функція для перевірки правильності формату рядка
def check_input_string(input_string):
    pattern = r'^\s*\d+\.\d+,\s*[A-Za-z\s]+\s*,\s*[A-Za-z\s]+\s*$'
    return bool(re.match(pattern, input_string))

# Зчитуємо рядок(ки) з даними від користувача та перевіряємо їх формат
input_strings = []
while True:
    input_string = input("Введіть рядок у форматі 'число, рядок1, рядок2': ")
    if not input_string:
        break
    if check_input_string(input_string):
        input_strings.append(input_string.strip())
    else:
        print("Неправильний формат рядка. Будь ласка, спробуйте ще раз.")

# Виводимо отримані дані у відповідному форматі
separator = " | "
print(f"{'Число':>7} {separator} {'Рядок 1':^20} {separator} {'Рядок 2':<20}")
print("-" * 50)
for input_string in input_strings:
    parts = input_string.split(",")
    number = float(parts[0].strip())
    string1 = parts[1].strip()
    string2 = parts[2].strip().upper()
    print(f"{number:>7.2f} {separator} {string1:^20} {separator} {string2:<20}")
```

## Завдання 6
Попросити користувача ввести цифри з клавіатури.

Перевірити, чи він виконав вимоги.

Якщо все вірно, то за допомогою форматування рядка вивести число та число помножене на 1.5:

"Користувач ввів: .. після множення на 1.5 результат: .."

```python
#Місце для коду Завдання 6
# Запитуємо у користувача ввід цифр
user_input = input("Будь ласка, введіть цифри: ")

# Перевіряємо, чи він виконав вимоги
if user_input.isdigit():
    # Якщо вірно, то конвертуємо рядок в число
    number = int(user_input)
    # Робимо обчислення
    result = number * 1.5
    # Виводимо результат з відповідним форматуванням
    print(f"Користувач ввів: {number}, після множення на 1.5 результат: {result}")
else:
    # Якщо невірно, виводимо повідомлення про помилку
    print("Помилка! Ви ввели не цифри.")
```
